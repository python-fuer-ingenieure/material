`begin golden_ratio`

```python
import sympy as sp
Phi, a, b = sp.symbols('Phi, a, b')
# Gleichung a/b = (a + b)/a definieren
# In SymPy sind wird die rechte Seite immer als = 0 angenommen
# Wir müssen also in diese Form umstellen
ab_eq = a/b - (a+b)/a
```

```python
ab_eq = ab_eq.expand()  # Summanden alle einzeln
```

```python
# Ersetze a/b durch Phi
phi_eq = ab_eq.subs([(a/b, Phi), (b/a, 1/Phi)])
```

```python
# Löse nach Phi
symb_solutions = sp.solve(phi_eq, Phi)
```

```python
num_solutions = [sol.evalf() for sol in symb_solutions]
```

`end golden_ratio`

`begin fourier_def --input_prefix=""`

```python
# Ohne "real=True" wird Integral nicht ausgewertet
# oder Fallunterscheidung komplizierter
def fourier_transform(s, t):
    # f > 0 ist streng genommen nicht sauber,
    # aber liefert lesbarere Ergebnisse
    f = sp.Symbol('f', real=True, positive=True)
    return sp.integrate(s*sp.exp(-2*sp.pi*sp.I*f*t),
                        (t, -sp.oo, sp.oo))
```

`end`

!! In der folgenden Zelle ist ein Tippfehler im Buch:

Statt `fourier_transform(rect(t)).simplify()` muss es `fourier_transform(rect(t), t).simplify()` heißen (2. Argument hat gefehlt).

`begin fourier_call`

```python
t = sp.Symbol('t', real=True)
fourier_transform(rect(t), t).simplify()
```

```python
# "Heaviside" ist Einheitssprung
#!fourier_transform(sp.exp(-t)*sp.Heaviside(t), t)
fourier_transform(sp.exp(-t)*sp.Heaviside(t), t).simplify() #!
```

`end`

# Lambdify

`begin lambdify_1`

```python
# Beispielfunktion: Taylor-Polynom von exp(-x)
# sp.series macht Reihenentwicklung
sp.series(sp.exp(-x))
```

```python
# removeO entfernt Landau-Symbol O
f_symb = sp.series(sp.exp(-x)).removeO()
```

```python
f_symb.subs(x, 1).evalf()
```

```python
#_
f_num = sp.lambdify(x, f_symb)
```

```python
f_num(1)
```

```python
#_
import numpy as np
xs = np.linspace(0, 5, 11)
```

```python
f_num(xs)  # Viele Auswertungen auf einmal
```

`end`

`begin lambdify_mat_1`

```python
x1, x2 = sp.symbols('x1, x2')
f_symb = sp.Matrix([[x1**2, sp.sin(x1)*sp.sin(x2)],
                    [sp.cos(x2), x2**2]])
```

```python
# Rückgabewert ist Funktionsobjekt
f_num = sp.lambdify([x1, x2], f_symb)
```

```python
f_num(1, 2)
```

`end`

`begin lambdify_mat_2`

```python
def f_vec_input(x):
    return f_num(*x)
x_vec = np.array([1, 2])
f_vec_input(x_vec)
```
